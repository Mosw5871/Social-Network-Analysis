# Networkx基础练习

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create an empty undirected graph
g = nx.Graph()
```

```python
# Add a single node
g.add_node('1')
```

```python
# Add multiple nodes from a list
g.add_nodes_from(['2', '3', 'four', '5'])
```

```python
# Add a single edge
g.add_edge('1', '2')
```

```python
# Add multiple edges from a list of tuples
g.add_edges_from([('1', '3'), ('1', 'four'), ('3', 'four')])
```

```python
# Add an edge with a weight attribute
g.add_edges_from([('2', '3', {"weight": 0.5})])
```

```python
# Add an edge with a different weight attribute
g.add_edges_from([('1', '2', {"weight": 0.9})])
```

```python
# Print all nodes in the graph
print(list(g.nodes()))
```

```python
# Print all edges in the graph
print(list(g.edges()))
```

```python
# Print neighbors of node '1'
print(list(g.neighbors('1')))
```

```python
# Print the degree of node '1'
print(g.degree('1'))
```

```python
# Print properties of node '1'
print(g['1'])
```

```python
# Print properties of the edge between '1' and '2'
print(g['1']['2'])
```

```python
# Print properties of all edges
print(g.edges.data())
```

```python
# Calculate and print the degree of each node in graph g
degrees = dict(g.degree())
print(degrees)
```

```python
# Calculate and print the clustering coefficient of each node in graph g
clustering = nx.clustering(g)
print(clustering)
```

```python
# Draw graph g with labels and bold fonts
nx.draw(g, with_labels=True, font_weight='bold')
plt.show()
```

```python
# Define color constants for graph visualization
GREEN = "#77DD77"
BLUE = "#99CCFF"
```

```python
# Draw graph k with specific colors, labels, and bold fonts
nx.draw(g, node_color=GREEN, with_labels=True, font_weight='bold', node_size=200, edge_color=BLUE)
plt.show()
```

```python
# Network Centrality (the higher result is better!)
# Degree: number of edges for node X
# Betweenness: number of shortest paths that pass through node X
# Closeness: average of the shortest paths between X and all other nodes
```

```python
# Calculate and print the degree centrality of each node in graph g
centrality = nx.degree_centrality(g)
print(centrality)
```

```python
# Calculate and print the betweenness centrality of each node in graph g
betweenness = nx.betweenness_centrality(g, normalized=True)
print(betweenness)
```

```python
# Calculate and print the closeness centrality of each node in graph g
closeness = nx.closeness_centrality(g)
print(closeness)
```

# 数据可视化实战

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取csv文件
data = pd.read_csv('/Users/shiwei/Desktop/上课/2024 秋季学期/大数据与社交网络/代码/some_class.csv', index_col=0)
```

```python
# 创建一个图
G = nx.Graph()
```

```python
# 添加节点
nodes = data.index
G.add_nodes_from(nodes)

# 添加边及权重
for i in range(len(data)):
    for j in range(i + 1, len(data)):
        weight = data.iloc[i, j]
        if weight!= 0:
            G.add_edge(nodes[i], nodes[j], weight=weight)
```

```python
# 基础网络分析
print("网络节点数：", G.number_of_nodes())
print("网络边数：", G.number_of_edges())
print("平均度数：", sum(dict(G.degree()).values()) / G.number_of_nodes())
```

```python
# 计算和展示基础的网络指标
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
```
